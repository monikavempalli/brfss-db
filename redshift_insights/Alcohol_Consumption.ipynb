```redshift
SELECT 
    income, 
    general_health, 
    COUNT(CASE WHEN alcohol_consumption != 'None' THEN 1 END) AS count_of_alcohol_consumers
FROM 
    brfss.lifestyle_factors
INNER JOIN 
    brfss.health_conditions 
ON 
    brfss.lifestyle_factors.respondent_id = brfss.health_conditions.respondent_id
INNER JOIN 
    brfss.respondents 
ON 
    brfss.lifestyle_factors.respondent_id = brfss.respondents.id 
WHERE 
    alcohol_consumption IS NOT NULL 
    AND income IS NOT NULL 
    AND general_health IN ('Poor','Fair','Good','Not Sure','Excellent', 'Very good')
GROUP BY 
    income, 
    general_health
Order BY
    income
```
