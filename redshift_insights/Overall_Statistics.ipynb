```redshift
--Getting the number of respondents with each chronic condition from all the 21 states:
SELECT 
    SUM(CASE WHEN diabetes = 'Yes' THEN 1 ELSE 0 END) AS num_diabetes,
    SUM(CASE WHEN asthma = 'Yes' THEN 1 ELSE 0 END) AS num_asthma,
    SUM(CASE WHEN high_blood_pressure = 'Yes' THEN 1 ELSE 0 END) AS num_high_blood_pressure,
    SUM(CASE WHEN high_cholesterol = 'Yes' THEN 1 ELSE 0 END) AS num_high_cholesterol,
    SUM(CASE WHEN heart_disease = 'Yes' THEN 1 ELSE 0 END) AS num_heart_disease,
    SUM(CASE WHEN stroke = 'Yes' THEN 1 ELSE 0 END) AS num_stroke
FROM brfss.chronic_condition;
```
