```redshift
--Getting the percentage of respondents with each chronic condition by state:
SELECT 
    state,
    100 * SUM(CASE WHEN diabetes = 'Yes' THEN 1 ELSE 0 END) / COUNT(*) AS percent_diabetes,
    100 * SUM(CASE WHEN asthma = 'Yes' THEN 1 ELSE 0 END) / COUNT(*) AS percent_asthma,
    100 * SUM(CASE WHEN high_blood_pressure = 'Yes' THEN 1 ELSE 0 END) / COUNT(*) AS percent_high_blood_pressure,
    100 * SUM(CASE WHEN high_cholesterol = 'Yes' THEN 1 ELSE 0 END) / COUNT(*) AS percent_high_cholesterol,
    100 * SUM(CASE WHEN heart_disease = 'Yes' THEN 1 ELSE 0 END) / COUNT(*) AS percent_heart_disease,
    100 * SUM(CASE WHEN stroke = 'Yes' THEN 1 ELSE 0 END) / COUNT(*) AS percent_stroke
FROM brfss.chronic_condition
JOIN brfss.respondents ON brfss.chronic_condition.respondent_id = respondents.id
GROUP BY state;
```
