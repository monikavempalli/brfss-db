```redshift
SELECT 
  brfss_redshift.brfss.respondents.state,
  AVG(brfss_redshift.brfss.health_conditions.physical_health) AS avg_physical_health,
  AVG(brfss_redshift.brfss.health_conditions.mental_health) AS avg_mental_health
FROM brfss_redshift.brfss.health_conditions
JOIN brfss_redshift.brfss.respondents ON brfss_redshift.brfss.health_conditions.respondent_id = brfss_redshift.brfss.respondents.id
GROUP BY brfss_redshift.brfss.respondents.state;
```
