```redshift
-- Finding the average physical health score by race
SELECT 
    r.race,
    AVG(hc.physical_health) AS avg_physical_health_score
FROM brfss_redshift.brfss.respondents r
JOIN brfss_redshift.brfss.health_conditions hc ON r.id = hc.respondent_id
GROUP BY r.race
ORDER BY avg_physical_health_score DESC;

-- Finding the average mental health score by race
SELECT 
    r.race,
    AVG(hc.mental_health) AS avg_mental_health_score
FROM brfss_redshift.brfss.respondents r
JOIN brfss_redshift.brfss.health_conditions hc ON r.id = hc.respondent_id
GROUP BY r.race
ORDER BY avg_mental_health_score DESC;
```
